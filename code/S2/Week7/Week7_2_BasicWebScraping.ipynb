```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://lifestyle.campus-star.com/scoop/6260.html"
```

```python
web_data = requests.get(url)
```

```python
print(web_data.text)
```

```python
soup = BeautifulSoup(web_data.text, 'html.parser')
```

```python
find_word = soup.find_all("p")
```

```python
for line in find_word:
  print(line)
```

```python
import re
clean = re.compile('<.*?>')

for line in find_word:
  clean_text = re.sub(clean, '', str(line))
  print(clean_text)
```

```python
#data collection 
clean = re.compile('<.*?>')
weird_name = []
for line in find_word:
  clean_text = re.sub(clean, '', str(line))
  weird_name.append(clean_text)
```

```python
len(weird_name)
```

```python
weird_name
```

```python
weird_name = weird_name[1:-2]
```

```python
for names in weird_name:
  e = names.split('\n')
  print(e)
```

```python
list_name = []
for names in weird_name:
  e = names.split('\n')
  for name in e:
    print(name)
    list_name.append(name)
```
